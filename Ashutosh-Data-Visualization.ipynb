#### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read CSV

```python
df = pd.read_csv('rideshare_kaggle.csv', index_col=0)
```

#### Displaying Head

```python
df.head()
```

#### Data Cleaning . Clearing All NULL Values

```python
df = df.dropna()
#df = df.sample(n=10000)
```

#### Scatter Plot of Price and Distance in accordance to Surge Multiplier (and remove the default regression line and tweak the graph limit with matplotlib)

```python
sns.lmplot(x='distance', y='price', data=df,
           fit_reg=False,
           hue='surge_multiplier')
plt.xlim(0,8.5)
plt.ylim(0,110)
```

#### Distance-Price Graph in Accordance to Surge Multiplier

```python
#sns.set(rc={'figure.figsize':(11.7,20)})
colors = ['#009999','#00994C','#2e856e','#4C9900','#999900','#f46d75','#990000']

sns.relplot(x='distance', y='price', data=df, kind ='scatter',col_wrap=3,
            col='surge_multiplier',hue='surge_multiplier', palette=colors).set(title="Distance-Price graph in accordance to Surge Multiplier")

plt.xlim(0,8.5)
plt.ylim(0,110)
```

#### 3D ploting the price-distance with surge_multiplier as z-axis and using a sample size of 10000

```python
from mpl_toolkits.mplot3d import Axes3D
sns.set(style = "darkgrid")

new_df=df.sample(n=10000)
x = new_df['distance']
y = new_df['price']
z = new_df['surge_multiplier']

fig = plt.figure()
fig.set_size_inches(12, 12)
ax = plt.axes(projection ="3d")

ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)

my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(x, y, z,
                    alpha = 0.8,
                    c=(x+y+z),
                    cmap = my_cmap,
                    marker ='^')

plt.title("Distance-Price graph in accordance to Surge Multiplier")
ax.set_xlabel("Distance",fontweight='bold')
ax.set_ylabel("Price",fontweight='bold')
ax.set_zlabel("Surge Multiplier",fontweight='bold')

ax.scatter(x, y, z)

plt.show()
```

#### Count of the number of cabs booked by LYFT and Uber

```python
colors=['#FF00BF','#1fbad6']     #LYft and Uber colors
plt.title("Booked Cabs")
ax = sns.countplot(x='cab_type', data=df,palette=colors)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[0])
```

#### Cabs Boarded from a specific source location and which cab was chosen

```python
plt.figure(figsize=(25,12))
colors=['#FF00BF','#1fbad6']
plt.title("Cabs boarded from source locations")
ax = sns.countplot(x='source',hue='cab_type', data=df,palette=colors)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

#### Cabs Boarded for a specific destination location and which cab was chosen

```python
plt.figure(figsize=(25,12))
plt.title("Cabs boarded to destination locations")
ax = sns.countplot(x='destination',hue='cab_type', data=df,palette=colors)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

#### Shows the type of cab people prefer to use 

```python
plt.figure(figsize=(25,12))
plt.title("The type of Cab chosen")
ax = sns.countplot(x='name',hue='cab_type', data=df,palette=colors)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```
